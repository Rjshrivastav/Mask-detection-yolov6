⚙️ Configuration

The dataset config file requires

1. The dataset root directory path and relative paths to train / val / test image directories and labels directories
2. The number of classes nc and
3. A list of class names:['With mask', 'Without mask']

```python
import yaml
import os

cwd = '/kaggle/working/mask_data/images'

data = dict(
    train =  os.path.join( cwd , 'train') ,
    val   =  os.path.join( cwd , 'val' ),
    is_coco = False,
    nc    = 2,
    names = ['With mask', 'Without mask'],
    )

with open(os.path.join( './YOLOv6/data', 'mask.yaml'), 'w') as outfile:
    yaml.dump(data, outfile, default_flow_style=False)

f = open(os.path.join( './YOLOv6/data' , 'mask.yaml'), 'r')
print('\nyaml:')
print(f.read())
```
